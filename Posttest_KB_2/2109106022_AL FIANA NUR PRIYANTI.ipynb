### Import library pandas dan numpy

pd.read_csv() untuk mengimport dataset file CSV ke pandas dataframe

```python
import pandas as pd
import numpy as np

df = pd.read_csv("googleplaystore.csv")
```

### Ketentuan 1 Analisis Dataset
*Informasi yang didapatkan*

df.shape[] : fungsi untuk menampilkan ukuran baris(record) atau kolom(atribut) dari suatu dataframe <br>
df.columns : fungsi untuk menampilkan nama-nama atribut atau kolom pada dataframe <br>
df.select_dtypes() : fungsi untuk mengembalikan dataframe baru yang menyertakan kolom dari tipe yang ditentukan <br>
include = object : untuk memilih semua tipe object <br>
include = np.number : untuk memilih semua tipe numerik/angka <br>
df.isna().any() : untuk memeriksa apakah ada atribut yang memiliki nilai null <br>
df.loc : untuk menyeleksi data pada lokasi tertentu

```python
record = df.shape[0]
atribut = df.columns
objek = df.select_dtypes(include=object).columns
angka = df.select_dtypes(include=np.number).columns
Null = df.loc[:, df.isna().any()].columns
```

### Ketentuan 2 Analisis Dataset
*Memasukkan informasi yang didapatkan ke dictionary dan menampilkannya*

mengakses dictionary dengan perulangan for
len() : fungsi untuk mengembalikan nilai panjang (jumlah anggota) dari suatu objek

```python
info = {
    'jumlah_record' : record,
    'nama_atribut' : atribut,
    'atribut_object' : objek,
    'atribut_angka' : angka,
    'atribut_null' : Null
}

print("""
================================================
        A N A L I S I S   D A T A S E T
================================================""")
print("Jumlah Record \t\t\t\t:", info['jumlah_record'])

print("================================================")
print("Jumlah Atribut \t\t\t\t:", len(info['nama_atribut']))
print("================================================")
print("------------------------")
print("  No\t| Nama Atribut")
print("------------------------")
for i in range (len(info['nama_atribut'])):
    print(" ", i, "\t|", info['nama_atribut'][i])
    i+=1
    
print("\n================================================")
print("Jumlah Atribut dengan Tipe Objek \t:", len(info['atribut_object']))
print("================================================")
print("-------------------------------")
print("  No\t| Atribut Tipe Object")
print("-------------------------------")
for i in range (len(info['atribut_object'])):
    print(" ", i, "\t|", info['atribut_object'][i])
    i+=1
    
print("\n================================================")
print("Jumlah Atribut dengan Tipe Angka \t:", len(info['atribut_angka']))
print("================================================")
print("------------------------------")
print("  No\t| Atribut Tipe Angka")
print("------------------------------")
for i in range (len(info['atribut_angka'])):
    print(" ", i, "\t|", info['atribut_angka'][i])
    i+=1
    
print("\n================================================")
print("Jumlah Atribut yang Memiliki Nilai Null :", len(info['atribut_null']))
print("================================================")
print("-------------------------------------")
print("  No\t| Atribut dengan Nilai Null")
print("-------------------------------------")
for i in range (len(info['atribut_null'])):
    print(" ", i, "\t|", info['atribut_null'][i])
    i+=1
    
print("\n================================================")
```
